## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv("data\hist_data.csv")
plt.hist(df["aデータ"], bins=30, alpha=0.5, label="aデータ", color="orange")
plt.hist(df["bデータ"], bins=30, alpha=0.5, label="bデータ", color="turquoise")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()
plt.title("aデータとbデータの温度に関する度数分布")
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv("data\scatter_data.csv")

class0 = df[df["class"] == 0]
class1 = df[df["class"] == 1]
class2 = df[df["class"] == 2]

plt.scatter(class0["length (cm)"],class0["width (cm)"], label="class0", color="red", marker="o")
plt.scatter(class1["length (cm)"],class1["width (cm)"], label="class1", color="blue", marker="o")
plt.scatter(class2["length (cm)"],class2["width (cm)"], label="class2", color="green", marker="o")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.legend()
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.grid()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv("data\line_data.csv")
plt.plot(df["期間"], df["件数"], color="steelblue", marker="o")
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")
plt.xticks(rotation=30)
plt.grid()
plt.show()
```

```python
df = pd.read_csv("data\line_data.csv")
plt.bar(df["期間"], df["費用"], color="steelblue")
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.title("期間ごとの費用の推移")
plt.xticks(rotation=30)
plt.grid()
plt.show()
```
